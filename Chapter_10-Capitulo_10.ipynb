### Caixa de código 13: Funções de onda do elétron em uma caixa unidimensional

```python
import numpy as np
import matplotlib.pyplot as plt

L = 50.0e-9                              # tamanho da caixa, em m
a0 = 5.2917721067e-11                        # raio de Bohr, em m
L_a0 = L / a0            # tamanho da caixa, em unidades atômicas

me = 9.1e-31                            # massa do elétron, em kg
xmin, xmax = 0, L_a0                            # bordas da caixa
npt = 1000                           # número de pontos da função
x = np.linspace(xmin, xmax, npt)                # grade de pontos

A = np.sqrt(2/L_a0)
psi_n1 = A * np.sin(1 * np.pi * x / L_a0)     # psi(x) para n = 1
psi_n2 = A * np.sin(2 * np.pi * x / L_a0)     # psi(x) para n = 2
psi_n3 = A * np.sin(3 * np.pi * x / L_a0)     # psi(x) para n = 3
psi_n4 = A * np.sin(4 * np.pi * x / L_a0)     # psi(x) para n = 4

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)  

plt.plot(x, psi_n1, '-y', label = 'n = 1')               # plot 1
plt.plot(x, psi_n2, '-g', label = 'n = 2')               # plot 2
plt.plot(x, psi_n3, '-r', label = 'n = 3')               # plot 3
plt.plot(x, psi_n4, '-b', label = 'n = 4')               # plot 4
plt.plot((0,xmax),(0,0),'--')        # linha horizontal tracejada
plt.legend(loc = 3, frameon=False)                      # legenda

plt.title('$\Psi_{n}(x)$ da partícula na caixa 1-D',fontsize=16)
plt.xlabel('x (em Bohr)',fontsize=14)
plt.ylabel('$\Psi_{n}(x)$',fontsize=14)
plt.tight_layout()
plt.savefig('particle_in_box_1D.png',dpi=300, transparent=True)
```

### Caixa de código 14: Usando o recurso de subgráﬁcos

```python
import numpy as np
import matplotlib.pyplot as plt

L = 50.0e-9                              # tamanho da caixa, em m
a0 = 5.2917721067e-11                        # raio de Bohr, em m
L_a0 = L / a0            # tamanho da caixa, em unidades atômicas   

me = 9.1e-31                            # massa do elétron, em kg
xmin, xmax = 0, L_a0                            # bordas da caixa
npt = 1000                           # número de pontos da função
x = np.linspace(xmin, xmax, npt)                # grade de pontos

A = np.sqrt(2/L_a0)
psi_n1 = A * np.sin(1 * np.pi * x / L_a0)     # psi(x) para n = 1
psi_n2 = A * np.sin(2 * np.pi * x / L_a0)     # psi(x) para n = 2
psi_n3 = A * np.sin(3 * np.pi * x / L_a0)     # psi(x) para n = 3
psi_n4 = A * np.sin(4 * np.pi * x / L_a0)     # psi(x) para n = 4

fig = plt.figure()

plt.subplot(2, 2, 1)
plt.plot(x, psi_n4,'-b', lw = 2.0)
plt.xlabel('x (em Bohr)')
plt.ylabel('$\Psi_{4}(x)$')
plt.plot((0,xmax),(0,0),'--')       # linha horizontal tracejada

plt.subplot(2, 2, 2)
plt.plot(x, psi_n3,'-r', lw = 2.0)
plt.xlabel('x (em Bohr)')
plt.ylabel('$\Psi_{3}(x)$')
plt.plot((0,xmax),(0,0),'--')       # linha horizontal tracejada

plt.subplot(2, 2, 3)
plt.plot(x, psi_n2, '-g', lw = 2.0)
plt.xlabel('x (em Bohr)')
plt.ylabel('$\Psi_{2}(x)$')
plt.plot((0,xmax),(0,0),'--')       # linha horizontal tracejada

plt.subplot(2, 2, 4)
plt.plot(x, psi_n1, '-y', lw = 2.0)
plt.xlabel('x (em Bohr)')
plt.ylabel('$\Psi_{1}(x)$')

plt.tight_layout()
plt.savefig('particle_in_box_1D_2.png',dpi=400,transparent=True)
```

### Caixa de código 15: Exemplo para plotar gráﬁcos logarítmicos

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0.1, 100.0, 0.1)
c = 1/10
y = np.exp(-c*x)
plt.figure(figsize = (8,6))
plt.subplot(2, 2, 1)
plt.plot(x, y, '-b', lw = 2.0)
plt.title('Escalas lineares em X e Y')
plt.xlabel("X", fontsize=10)
plt.ylabel("Y", fontsize=10)
plt.grid()
plt.subplot(2, 2, 2)
plt.semilogy(x, y, '-r', lw = 2.0)
plt.xlabel("X", fontsize = 10)
plt.ylabel("log(Y)", fontsize = 10)
plt.title('Escala linear em X e logarítmica em Y')
plt.grid()
plt.subplot(2, 2, 3)
plt.semilogx(x, y,'-g', lw = 2.0)
plt.xlabel("log(X)", fontsize = 10)
plt.ylabel("Y", fontsize = 10)
plt.title('Escala linear em Y e logarítmica em X')
plt.grid()
plt.subplot(2, 2, 4)
plt.loglog(x, y, '-m', lw = 2.0)
plt.xlabel("log(X)", fontsize = 10)
plt.ylabel("log(Y)", fontsize = 10)
plt.title('Escalas logarítmicas em X e Y')
plt.grid()
plt.tight_layout()
plt.savefig('escalas_log.png', dpi=400, transparent=True)
```

### Caixa de código 18: Funções angulares de orbitais hidrogenóides 3d

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.arange(0, 2*np.pi, 0.01)

P_2_2 = abs((np.sqrt(15)/4) * (np.sin(theta)**2))
P_2_1 = abs((np.sqrt(15)/2) * (np.sin(theta)*np.cos(theta)))
P_2_0 = abs((np.sqrt(10)/4) * (3*np.cos(theta)**2 - 1))

fig = plt.figure(figsize=(12,12))

ax = plt.subplot(221, projection='polar')
ax.plot(theta, P_2_0, color = 'b')
ax.set_rmax(1.8)
ax.set_rticks([0.4, 0.8, 1.2, 1.6])  
ax.set_title(label='$P(\Theta)_{3,0,0}$', fontsize=16)
ax.grid(True)

ax = plt.subplot(222, projection='polar')
ax.plot(theta, P_2_1, color = 'r')
ax.set_rmax(1.2)
ax.set_rticks([0.3, 0.6, 0.9, 1.2])  
ax.set_title(label='$P(\Theta)_{3,2,\pm1}$', fontsize=16)
ax.grid(True)

ax = plt.subplot(223, projection='polar')
ax.plot(theta, P_2_2, color = 'g')
ax.set_rmax(1.2)
ax.set_rticks([0.3, 0.6, 0.9, 1.2])  
ax.set_title(label='$P(\Theta)_{3,2,\pm2}$', fontsize=16)
ax.grid(True)
plt.tight_layout()
plt.savefig('polar_orbitais_d.png', dpi=400, transparent=True)
```

### Caixa de código 22: Gráﬁco de barras empilhadas com percentagens

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

cores = ('b', 'y', 'g', 'r')
AA = ['Ala','Gly','Tyr','Pro']
atomos = ['C','H', 'N', 'O']
C = [40.4432,32.0001,59.6596,52.1613]
H = [7.9193,6.7136,6.1192,7.8793]
N = [15.7214,18.6590,7.7305,12.1659]
O = [35.9161,42.6272,26.4907,27.7935]
x = np.arange(len(AA))    
L = 0.75
data = (C, H, N, O)
piso = np.zeros(len(AA)) 

# Barras empilhadas com texto de percentagens anotadas 

for y, cor in zip(data, cores):
    plt.bar(x, y, L, bottom=piso, color=cor, edgecolor='k', 
            label = str(atomos[cores.index(cor)]))
    for i in x:
        plt.text(x[i], piso[i] + y[i]/2., str("%.1f"%y[i]), ha="center", 
        va="center",color="white", fontsize=12, fontweight="bold")
    piso += y    
    
plt.ylim(0,110)
plt.xticks(x, AA, fontweight='bold', fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel("Aminoácido", fontsize=16, labelpad = 12)
plt.ylabel("Composição em Massa (%)", fontsize=16, labelpad = 10)

plt.legend(bbox_to_anchor=(1.01,0.5), loc='center left', fontsize=12)
plt.tight_layout()
plt.savefig('stacked_bars_with_percent.png',dpi=400, transparent=True)
```

### Caixa de código 25: Histograma da concentração de Cd<sup>2+</sup> em amostras de água

```python
import numpy as np
import matplotlib.pyplot as plt

conc_Cd = [31.5, 43.3, 39.8, 41.7, 42.4, 53.4, 40.5, 42.9, 42.4, 43.9, 53.5, 19.3, 
           44.7, 30.7, 39.9, 58.3, 44.6, 27.2, 36.5, 31.3]

med = np.mean(conc_Cd)
std = np.std(conc_Cd)
menor = np.min(conc_Cd)
maior = np.max(conc_Cd)
npt = len(conc_Cd)
faixa = maior - menor
num_interv = int(np.sqrt(npt))
largura =  faixa / num_interv 

x = np.arange(menor, maior, 0.1) 
y = (1/(std*np.sqrt(2*np.pi)) * np.exp(-(x - med)**2 / (2*std**2)))
plt.hist(conc_Cd, bins=5, color='b', edgecolor='w', density=True)
plt.plot(x, y, linewidth=2, color='r')
plt.xlabel('Concentração de íons $Cd^{2+}$ (em $\mu$g/L)', fontsize=14)
plt.ylabel('Probabilidade',fontsize=14)

plt.tight_layout()
plt.savefig('histograma_Cd+2.png', dpi=400, transparent=True)
```

### Caixa de código 28: Exemplo para plotar gráﬁcos 3D com pontos

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x = np.arange(0, 10, 1)
y = np.arange(0, 10, 1)

X, Y = np.meshgrid(x, y)

Z = 1 - X**2 + Y**2

fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(X, Y, Z, c = Z)

ax.set_xlabel('Eixo X', fontsize = 14, labelpad = 12)
ax.set_ylabel('Eixo Y', fontsize = 14, labelpad = 12)
ax.set_zlabel(r'Z(X,Y)=1-$X^{2}-Y^{2}$',fontsize=14,labelpad=12)

ax.set_title(label='Gráfico 3D com pontos', fontsize = 16)
ax.grid(True)

plt.savefig('3D_plot_with_points.png', dpi=400, transparent=True)
```

### Caixa de código 29: Parabolóide hiperbólico: função Z(X,Y) = X<sup>2</sup> - Y<sup>2</sup>

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-1,1,1000)
y = np.linspace(-1,1,1000)
X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2

ax.plot_surface(X, Y, Z, linewidth=0, antialiased=False, shade = True, alpha = 0.5)
ax.set_xlabel('X', fontsize=18, labelpad=12)
ax.set_ylabel('Y', fontsize=18, labelpad=12)
ax.set_zlabel('$Z(X,Y) = X^2 - Y^2$', fontsize=18, labelpad=12)

plt.xticks([-1,-0.5,0,0.5,1], fontsize = 14)
plt.yticks([-1,-0.5,0,0.5,1], fontsize = 14)
ax.set_zticks([-1,-0.5,0,0.5,1]) # eixo Z: formatação diferente!
for p in ax.zaxis.get_major_ticks(): p.label.set_fontsize(14)
plt.tight_layout()
plt.savefig('3D_surface_plot_01.png', dpi=400, transparent=True)
```

### Caixa de código 31: Função de onda da Partícula na caixa 2D, Ψ<sub>2,2</sub>(X, Y )

```python
import numpy as np
import matplotlib.pyplot as plt

L, nx, ny = 1, 2, 2
x = np.linspace(0, L, 20)
y = np.linspace(0, L, 20)
X, Y = np.meshgrid(x,y)

Psi = (2/L) * np.sin(nx*np.pi*X/L) * np.sin(ny*np.pi*Y/L)

fig = plt.figure(figsize=(10,6))
plt.contourf(X, Y, Psi, 12, cmap='jet', alpha=0.8)
contorno = plt.contour(X, Y, Psi, 12, colors='black')
plt.clabel(contorno)

plt.title('Gráfico 2D de $\Psi_{2,2}(X,Y)$', fontsize=16)
plt.ylabel('Y', fontsize = 14)
plt.xlabel('X', fontsize = 14)
plt.savefig('2D_part_in_box_600.png', dpi=600, transparent=True)
```

```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri

arq = "dados_scan_2-buteno.csv"
D1,D2,E = np.loadtxt(arq, skiprows=1, delimiter=',', unpack=True)

for i in range(D1.shape[0]):
    if D1[i] > 360:
        D1[i] -= 360
    if D2[i] > 360:
        D2[i] -= 360

fig, ax1 = plt.subplots()
ax1.tricontour(D1,D2,E, levels=15, linewidths=0.5, colors = 'k')
cntr = ax1.tricontourf(D1, D2, E, levels = 15, cmap = "jet")
fig.colorbar(cntr, ax = ax1)

plt.xlabel('D1 ($^\circ$)', fontsize=14)
plt.ylabel('D2 ($^\circ$)', fontsize=14)
plt.title('Análise conformacional do 2-Buteno\n', fontsize = 16)
plt.tight_layout()
plt.savefig('3D_plot_conform_OK.png', dpi=400, transparent=True)
```
